```python
import numpy as np
import scipy.linalg as la

def genQ(i, j):
    '''
    Calculates de (i,j) entry of the generator matrix, qij
    '''
    if i == 0 and j == 0:
        qij = -1.5
    elif i == 0 and j == 1:
        qij = 1.5
    else:
        if i < 5:
            if i-1 == j:
                qij = 1
            elif i == j:
                qij = -2.5
            elif i + 1 == j:
                qij = 1.5
            else:
                qij = 0
        else:
            if i-1 == j:
                qij = 2
            elif i == j:
                qij = -3.5
            elif i+1 == j:
                qij = 1.5
            else:
                qij = 0
    
    return(qij)
```

```python
N = 6

Q = np.zeros([N+1, N+1])

for i in range(N+1):
    for j in range(N+1):
        Q[i,j] = genQ(i, j)
        
print(Q)
```

```python
N = 4

Q = np.zeros([N+1, N+1])

for i in range(N+1):
    for j in range(N+1):
        Q[i,j] = genQ(i, j)
        
print(Q)

P, L, U = la.lu(Q)
print('This is the permutation:')
print(P)
```

```python
N = 25

Q = np.zeros([N+1, N+1])

for i in range(N+1):
    for j in range(N+1):
        Q[i,j] = genQ(i, j)
        
print(Q)

P, L, U = la.lu(Q)
print('This is the permutation:')
print(P)
```

```python
print('With permutation')
print(Q*P)
print('L')
print(L)
print('U')
print(U)
```
